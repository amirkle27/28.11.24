```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv(r"C:\Users\123\Downloads\sport_athletes.csv")
display(df)
```

## עבור כל ענף ספורט- הצג את המשקל הגבוה ביותר , הגובה המקסימלי ומס' הזכיות המקסימליות 
### a. שנה את שם עמודה weight ל- weight_max
### b. שנה את שם עמודה height ל - height_max
### c. שנה את שם העמודה wins ל - wins_max

```
max_df = df.groupby('Sport Type')[['Athlete Name','Weight(kg)','Height(cm)','Wins']].max().rename(columns={'Weight(kg)':'weight_max','Height(cm)':'height_max','Wins':'wins_max'})
display(max_df)
```

##  עבור כל ענף ספורט- הצג את המשקל המינימלי, כעת מיין את ענפי הספורט לפי המשקל הנמוך בסדר עולה


```
min_df = df.groupby('Sport Type')[['Athlete Name','Weight(kg)']].min().sort_values(['Weight(kg)'],ascending=[True])
display(min_df)
```

## .3 מכל ענף ספורט - הצג את 2 המשתתפים הגבוהים ביותר


```
height_df = df.groupby('Sport Type')['Height(cm)'].nlargest(2)
display(height_df)
```

## .4 צור groups לפי מדינה וענף ספורט
### a. הצג את הסטטיסטיקות הגבוהות ביותר

### c. הצג את הסטטיסטיקות הגבוהות ביותר עבור כדורגל בלבד. רמז xs
### d. הצג את הסטטיסטיקות הגבוהות ביותר עבור טניס וכדורגל בלבד. רמז isin
### e. הפוך את האינדקסים לפי ספורט-מדינה ומיין את המידע לפי ספורט+מדינה
### רמז swaplevel
### f. הצג את הסטטיסטיקות ש ל ערך: מינימילי, מקסימלי, ממוצע וסטיית תקן ,
### עבור העמודות : מספר זכיות וגובה. רמז agg

```
display(df.groupby(['Country','Sport Type']).max().loc['USA','Basketball'])
```

```
display(df.groupby(['Country','Sport Type']).max().xs(level = 'Sport Type', key='Soccer'))
```

```
tennis_soccer=df[df['Sport Type'].isin(['Tennis','Soccer'])].groupby(['Country','Sport Type']).max()
display(tennis_soccer)
```

```
swapped_tennis_soccer = tennis_soccer.swaplevel('Country','Sport Type')
display(swapped_tennis_soccer)
```

```
display(df.groupby(['Country','Sport Type']).agg({'Height(cm)':['min','max','mean','std'],'Wins':['min','max','mean','std']})) 
```

## .5 הוסף את השורות הבאות לטבלה באמצעות concat:
### new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
### 'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
### 'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })


```
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })
```

```
new_df = pd.concat([df,new_rows],axis=0)
display(new_df)
```

## .6 הפוך את תאריך האולימפידה ל - datetime. רמז datetime_to,
### כעת הצג את כל הזכיות לפי שנים רמז rule resample,
### הוסף פילטר כך שרק השנים בהם יש זכיה )גדול מ - אפס( יופיעו

```
new_df['Olympic Date'] = pd.to_datetime(new_df['Olympic Date'],format = '%d/%m/%Y')
```

```
new_df.set_index('Olympic Date', inplace=True)
```

```
wins_by_year = new_df.resample(rule='YE')['Wins'].sum()
```

```
display(wins_by_year)
```

## *בונוס: הפוך את כל הערכים של 0 ל- NaN ואז השתמש ב - dropna

```
display(wins_by_year)
```

```
wins_by_year.replace(0, np.nan, inplace=True)
display(wins_by_year)
```

```
wins_by_year.dropna()
```
