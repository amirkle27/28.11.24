##  ◊¢◊ë◊ï◊® ◊õ◊ú ◊¢◊†◊£ ◊°◊§◊ï◊®◊ò- ◊î◊¶◊í ◊ê◊™ ◊î◊û◊©◊ß◊ú ◊î◊û◊ô◊†◊ô◊û◊ú◊ô, ◊õ◊¢◊™ ◊û◊ô◊ô◊ü ◊ê◊™ ◊¢◊†◊§◊ô ◊î◊°◊§◊ï◊®◊ò ◊ú◊§◊ô ◊î◊û◊©◊ß◊ú ◊î◊†◊û◊ï◊ö ◊ë◊°◊ì◊® ◊¢◊ï◊ú◊î


```
min_df = df.groupby('Sport Type')[['Athlete Name','Weight(kg)']].min().sort_values(['Weight(kg)'],ascending=[True])
display(min_df)
```
